```python
import pandas as pd
movie= pd.read_csv('movies.dat', delimiter='::', header=None, 
        names=['MovieID', 'name', 'genre'], engine='python')

Ratings = pd.read_csv("ratings.csv")
Tags = pd.read_csv("users.csv")
```

```python
Ratings.columns=["UserID","MovieID","Rating_x","timestamp"]

```

```python
mean=pd.DataFrame(Ratings.groupby("UserID",as_index=False)["Rating_x"].mean())
mean.rename(columns={"Rating_x":"Rating_y"})
```

```python
Ratings_avg=pd.merge(Ratings,mean,on='UserID')
Ratings_avg
```

```python
Ratings_avg["avg"]=Ratings_avg["Rating_x_x"]-Ratings_avg["Rating_x_y"]
```

```python
Ratings_avg
```

```python
from sklearn.metrics.pairwise import cosine_similarity
matrix=pd.pivot_table(Ratings_avg,values="avg",index="UserID",columns="MovieID")
```

```python
f_matrix=matrix.fillna(matrix.mean(axis=0))
```

```python
import numpy as np
cosine = cosine_similarity(f_matrix)
np.fill_diagonal(cosine,0)
user_m=pd.DataFrame(cosine,index=f_matrix.index,columns=f_matrix.index)
user_m
```

```python
a=pd.DataFrame(user_m.index,index=user_m.index)
n=5
for i in user_m.index:
    b=user_m[i]
    c=pd.DataFrame(b.sort_values(ascending=False))
    for j in range(n):
        
        a.at[i,"Top{}".format(j+1)]=c.index[j]
    
    
   
    
a
```

```python
def score(user,item):
    r_avg=mean["Rating_x"][user-1]
    s=0
    ws=0
    for i in range(n):
        ng=a["Top{}".format(i+1)][user]
        r=f_matrix[item][ng]
        w=user_m[ng][user]
        x=w*r
        s=s+x
        ws=ws+w
    return r_avg+(s/ws)
        
        
        
    
```

```python
def recommend_movie(user):
    already_watched=matrix.columns[matrix[matrix.index==user].notna().any()].tolist()
    movies_list=list()
    for i in range(n):
        g=a["Top{}".format(i+1)][user]
        k=Ratings_avg.loc[Ratings_avg["UserID"]==g].reset_index()
        for j in range(len(k)):
            movies_list.append(k["MovieID"][j])
    recommendable=list(set(movies_list)-set(already_watched))
    scores=[]
    for item in recommendable:
        scores.append(score(user,item))
    final=pd.DataFrame({"MovieID":recommendable,"scores":scores})
    top_5_movies= final.sort_values(by='scores',ascending=False).head(5)
    res=top_5_movies.merge(movie,on="MovieID")
    res=res.name.values.tolist()
    
    return res
```

```python
user = int(input("Enter the user id  : "))
predicted_movies = recommend_movie(user)
print(" ")
print("The Recommendations for User Id : 370")
print("   ")
for i in predicted_movies:
    print(i)
```
